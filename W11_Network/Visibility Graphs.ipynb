```python
%matplotlib widget
import numpy as np
import matplotlib.pyplot as plt
from ts2vg import NaturalVG
import networkx as nx
```

```python
t = np.linspace(0,25,250)
y = np.sin(t)
y1 = np.cos(2*t)
plt.figure()
#plt.plot(t,y)
#plt.plot(t,y1)

plt.plot(t,y+y1)
z=y+y1
```

```python
vg = NaturalVG()
vg.build(z)
ad = vg.adjacency_matrix()
nxg = vg.as_networkx()

plt.figure(figsize=(8,5))

plt.subplot(141)
plt.plot(t,z)


plt.subplot(142)
plt.imshow(ad,cmap='gray')

graph_plot_options = {
    "with_labels": False,
    "node_size": 2,
    "node_color": [(0, 0, 0, 1)],
    "edge_color": [(0, 0, 0, 0.15)],
}

plt.subplot(143)
nx.draw_networkx(nxg,pos=vg.node_positions(),**graph_plot_options)


plt.subplot(144)
nx.draw_networkx(nxg,pos=nx.kamada_kawai_layout(nxg),**graph_plot_options)

ks, ps = vg.degree_distribution

plt.figure()
plt.plot(ks,ps)
```
